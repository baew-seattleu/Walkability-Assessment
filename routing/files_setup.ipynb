How to obtain edges/nodes csv files


Also run this to get seattle_walk.graphml

```python
import osmnx as ox
import pandas as pd
# Download and save the graph
G_2 = ox.graph_from_place('Seattle, Washington, USA', network_type='walk')
ox.save_graphml(G_2, filepath="../data/seattle_walk.graphml")
```

```python
# Defining the map boundaries for Seattle Downtown Area
north, east, south, west = 47.62425802431359, -122.35919327219091, 47.596252480947555, -122.31765250756945
# Downloading the map as a graph object
graph = ox.graph_from_bbox(bbox=(north, south, east, west), network_type = 'walk')
```

```python
nodes, edges = ox.graph_to_gdfs(graph)
nodes.to_csv('../data/nodes.csv', index=True)
nodes.head()
edges.to_csv('../data/edges.csv', index=True)
edges.head()
```
