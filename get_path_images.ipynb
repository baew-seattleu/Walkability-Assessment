```python
import pandas as pd
from IPython.display import display, Image
import os
from PIL import Image as PILImage
```

```python
# Path to the CSV file
file_path = r"C:\Users\garla\first-node\WalkExplorer\routing\data\ex3\path_images.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Directory where images are stored
image_directory = r"C:\Users\garla\first-node\WalkExplorer\open_ai\SVI"
output_directory =r"C:\Users\garla\first-node\WalkExplorer\routing\data\ex3\saved_path_images"

# Ensure the output directory exists
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Iterate through each row in the DataFrame and display the path information and images
for index, row in df.iterrows():
    path_number = row['Path_Number']
    score_method = row['Score_Method']
    path_score = row['Path_Score']
    image_sequence = row['Image_Sequence'].split(', ')  # Split the images by comma and space

    # Display path number, score method, and path score
    print(f"Path Number: {path_number}")
    print(f"Score Method: {score_method}")
    print(f"Path Score: {path_score}")
    print("Images:")

 # Initialize image counter for each path. 
 # The image counter will keep the path images in sequential order when displayed in a folder
    image_counter = 0

    for image_name in image_sequence:
        if image_name:
            image_path = os.path.join(image_directory, image_name)
            if os.path.exists(image_path):
                # Load the image using PIL
                img = PILImage.open(image_path)
                # Convert the image to RGB to ensure compatibility with PNG format
                img = img.convert('RGB')  

                # Save the image to the output directory as PNG with path number and image counter
                saved_image_path = os.path.join(output_directory, f"{score_method}_{path_number}_{image_counter}_{image_name}.png")
                img.save(saved_image_path)

                # Increment the image counter
                image_counter += 1
    # Display each image in the sequence using Image() from IPython
    #for image_name in image_sequence:
    #    if image_name:  # Ensure the image name is not empty
    #        image_path = os.path.join(image_directory, image_name)
    #        if os.path.exists(image_path):
    #            display(Image(filename=image_path))  # Display the image inline
    #        else:
    #            print(f"Image {image_name} not found.")
```
